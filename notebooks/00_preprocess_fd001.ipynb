# This notebook converts the raw FD001 dataset into a single clean, dataset with computed RUL for use in EDA and modeling.

```python
# Import libraries for paths and data reading
import pandas as pd
from pathlib import Path
```

```python
RAW_PATH = Path('../data/raw/train_FD001.txt')

df = pd.read_csv(RAW_PATH, sep=r'\s+', header=None)
```

```python
# Label columns
columns = (
    ['unit_number', 'time_in_cycles']
    + [f'operation_setting_{i}' for i in range(1, 4)]
    + [f'sensor_{i}' for i in range(1, 22)]
)

# Assign column names to sensor readings
df.columns = columns
df
```

```python
# Compute max_cycle per engine
max_cycles = df.groupby('unit_number')['time_in_cycles'].transform('max')

# Compute RUL
df['RUL'] = max_cycles - df['time_in_cycles']
df
```

```python
PROCESSED_PATH = Path('../data/processed/fd001_processed.csv')
df.to_csv(PROCESSED_PATH, index=False)
```
